```python
from src.phishIRISData import PhishIRISDataset
from src.config import config
import src.SIFT_bovw as bovw
import numpy as np
```

```python
train_dataset = PhishIRISDataset(
    data_dir=config.PHISHIRIS_DL_DATASET_DIR, split="train", preprocess=False
)
val_dataset = PhishIRISDataset(
    data_dir=config.PHISHIRIS_DL_DATASET_DIR, split="val", preprocess=False
)
```

```python
bw_images = bovw.read_bw_images([img_path for img_path, _ in train_dataset])
```

```python
keypoints, descriptors = bovw.sift_features(bw_images)
```

```python
all_descriptors = []
# extract image descriptor lists
for img_descriptors in descriptors:
    # extract specific descriptors within the image
    for descriptor in img_descriptors:
        all_descriptors.append(descriptor)
# convert to single numpy array
all_descriptors = np.stack(all_descriptors)
all_descriptors.shape
```

```python
from scipy.cluster.vq import kmeans

k = 200
iters = 1
codebook, variance = kmeans(all_descriptors, k, iters)
```

```python
import joblib

# save number of clusters and codebook
# Joblib dumps Python object into one file
joblib.dump((k, codebook), config.PROCESSED_DATA_DIR / "bovw-train.pkl", compress=3)
```
