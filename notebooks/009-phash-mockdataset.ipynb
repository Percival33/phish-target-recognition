```python
from perception import hashers
import pandas as pd
from src.config import config
```

```python
df = pd.read_csv(config.RAW_DATA_DIR / "mockup" / "dataset.csv")
df
```

```python
df['screenshot_hash'][0]
```

```python
for hasher in [hashers.PHashF(), hashers.PHashU8(), hashers.DHash()]:
    try:
        print(f"Using hasher: {hasher} -> {hasher.string_to_vector(df['screenshot_hash'][0])}")
    except Exception as e:
        print(f"Error: {hasher} {e}")
```

```python
hasher = hashers.DHash()
hash_phish = df['screenshot_hash'][0]
hash_benign = hasher.compute(str(config.RAW_DATA_DIR / 'olx.png'))
hasher.compute_distance(hash_benign, hash_phish)
```

```python
for i in range(df.shape[0]):
    hash_phish = df['screenshot_hash'][i]
    print(hasher.compute_distance(hash_benign, hash_phish), f" -> {df['affected_entity'][i]}")
```

```python
for i in range(df.shape[0]):
    hash_phish = df['screenshot_hash'][i]
    print(hasher.compute_distance(df['screenshot_hash'][0], hash_phish), f" -> {df['affected_entity'][i]}")
```
