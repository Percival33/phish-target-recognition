# Phash-dbscan
`phash` used as image representation and clustering with `dbscan`

```python
from src.phishIRISData import PhishIRISDataset
from src.config import config
from sklearn.cluster import dbscan
import numpy as np
```

```python
train_dataset = PhishIRISDataset(
    data_dir=config.PHISHIRIS_DL_DATASET_DIR, split="train", preprocess=False
)
```

```python
core_samples, labels = dbscan(
    list(train_dataset.img_to_hash.values()), eps=3, min_samples=5
)
```

```python
np.count_nonzero(labels == -1), np.unique(labels, return_counts=True)
```
